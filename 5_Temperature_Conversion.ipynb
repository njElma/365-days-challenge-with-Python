# To convert temperature Fahrenheit to Celsius

# Celsius=((f-32)x5)/9




```python

f=float(input("f:"))
celsius=((f-32)*5)/9
print("Celsius:", celsius)
```

# To convert temperature Celsius to Fahrenheit  

# Fahrenheit=((9xc)/5)+32

```python
c=float(input("c:"))
fahrenheit=((9*c)/5)+32
print("fahrenheit:", fahrenheit)
```
