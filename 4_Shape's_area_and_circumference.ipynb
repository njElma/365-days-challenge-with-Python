# To calculate circle's area and circumference
1.   Area= π x r x r [r= Radius; π= 3.1416]
2.   Circumference= 2 x π x r



```python
pi=3.1416
radius=6
area=pi*radius*radius
circumference=2*pi*radius
print("Area: ",area)
print("Circumference:", circumference)
```

```python
# To take user input
pi=3.1416
radius=float(input("Enter radius:"))
area=pi*radius*radius
circumference=2*pi*radius
print("Area: ",area)
print("Circumference:", circumference)
```

# To calculate rectangle's area and circumference
1.   Area= length x width
2.   Circumference= 2 x (length + width)

```python
length=float(input("Enter length: "))
width=float(input("Enter width: "))
area=length*width
circumference=2*(length+width)
print("Area:", area)
print("circumference:", circumference)
```

# To calculate triangle's area
1.   Area= (base x height)/2


```python
base=float(input("base:"))
height=float(input("height:"))
area=base*height/2
print("Area:", area)
```

# To calculate parallelogram's area and circumference
**1.   Area= base x height**

**2. Circumference= 2 x (base + height)**

```python
base=float(input("base:"))
height=float(input("height:"))
area=base*height
circumference=(base+height)*2
print("Area:", area)
print("Circumference:", circumference )
```
